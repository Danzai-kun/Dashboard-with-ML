```python
#making a subset of the dataset
imbd_Movies = imbd[['Movie Name', 'Year of Release', 'Movie Rating','Watch Time']]
```

```python
#coversion of string data into integer format
imbd_Movies['Watch Time'] = imbd_Movies['Watch Time'].str.extract('(\d+)').astype(float)
```

```python
#Retrieval of required data
imbd_Movies[imbd_Movies['Movie Rating']>8.6]
```

```python
#Retrieval of some required data
imbd_Movies[imbd_Movies['Watch Time']> 190]
```

```python
#coversion of string data into integer format
imbd_Movies['Year of Release'] = imbd_Movies['Year of Release'].str.extract('(\d+)').astype(float)
```

```python
imbd_Movies.describe
```

```python
#Grouping data to be display of on the graph
imbd_info = imbd_Movies.groupby('Movie Name', as_index=False).agg({"Year of Release":"sum","Watch Time":"sum"})
```

```python
#Retrieval of some required data
imbd_info[imbd_info['Watch Time']> 220]
```

```python
#Grouping data to be display of on the graph
imbd_data = imbd_Movies.groupby('Movie Name', as_index=False).agg({"Year of Release":"sum","Movie Rating":"sum"})
```

```python
#Retrieval of some required data
imbd_data[imbd_data['Movie Rating']> 8.8]
```

```python
#creating graph and figures for visualization dashboard
fig , axes = plt.subplots(2,2, figsize = (12,12))
sns.set_style("darkgrid")

fig1 = sns.barplot(x = "Watch Time", y ="Movie Name",
                   data = imbd_info[imbd_info['Watch Time']> 220] , ax = axes[0,0])
fig1.set_title("Movies With Highest Watch Time")

fig2 = sns.barplot(x = "Movie Rating", y ="Movie Name",
                   data = imbd_data[imbd_data['Movie Rating']> 8.8] , ax = axes[1,0])
fig2.set_title("Movies with highest Rating")

fig3 = sns.histplot(imbd_info['Watch Time'], bins=5, kde=True, ax=axes[0, 1] , color = 'dodgerblue')
fig3.set_title('Histogram of Watch Time')

fig4 = sns.lineplot(x ="Movie Rating", y ="Year of Release",
                    data = imbd_data[imbd_data['Movie Rating']> 8.8], ax=axes[1, 1])
fig4.set_title("Movies Rating as per Release")
plt.show()
```

```python
#making a second subset of the dataset
imbd_Movieset = imbd[['Movie Name', 'Meatscore of movie', 'Votes','Gross','Movie Rating']]
```

```python
#coversion of string data into integer format
imbd_Movieset['Meatscore of movie'] = imbd_Movieset['Meatscore of movie'].str.extract('(\d+)').astype(float)
```

```python
#coversion of string data into integer format
imbd_Movieset['Gross'] = imbd_Movieset['Gross'].str.extract('(\d+)').astype(float)
```

```python
imbd_Movieset.head()
```

```python
#Grouping the required data
imbd_gross = imbd_Movieset.groupby('Movie Name', as_index=False).agg({"Votes":"sum","Gross":"sum"})
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
sns.set_style("darkgrid")

fig1 = sns.histplot(imbd_Movieset['Meatscore of movie'], bins=5, kde=True, ax=axes[0, 0],color='coral')
fig1.set_title('Movies Metascore')

# Subplot 2: Lineplot to show Gross income as per Metascore of Movie
fig2 = sns.lineplot(x='Meatscore of movie', y='Gross',
                       data = imbd_Movieset[imbd_Movieset['Gross']>360], ax=axes[1, 0], color = 'cyan')
fig2.set_title('Gross income as per Metascore of Movie')

#Subplot 3: stack plot of Gross for every rating
pivot_data = imbd_Movieset[imbd_Movieset['Movie Rating'] > 8.6].pivot_table(
    index='Meatscore of movie', values='Gross', aggfunc='sum', fill_value=0
)
colors = ['r']
axes[0, 1].stackplot(pivot_data.index, pivot_data['Gross'], labels=['Gross'], color = colors)
axes[0, 1].set_title('Gross Income as per Rating score')
axes[0, 1].legend(title='Gross')

# Subplot 4: Stair Plot of Movie Rating Distribution
rating_distribution = imbd_Movieset['Movie Rating'].value_counts().sort_index()
step_color = 'mediumseagreen'
axes[1, 1].step(rating_distribution.index, rating_distribution.values, where='mid', marker='o', color = step_color)
axes[1, 1].set_title('Movie Rating Distribution')


plt.tight_layout()
plt.show()
```
